```python
from fuzzywuzzy import fuzz
def func(df,column,ind,ratio):
    for sea in df[column] :
        percent =fuzz.ratio(ind.lower(), sea.lower())
        if percent>= ratio:
            rows_with_matches = df[column] == sea
            df.loc[rows_with_matches, column] = ind

func(df= professors, column= "Country", ind="south korea", ratio=47)
professors["Country"]
```
