# **dropna() vs dropna(axis=1):**


df.dropna(): This method removes rows containing any missing values in the DataFrame. If there is at least one NaN or None value in a row, that entire row will be dropped from the DataFrame. The resulting DataFrame will have all missing values removed along the row axis (axis=0).

df.dropna(axis=1): This method removes columns containing any missing values in the DataFrame. If there is at least one NaN or None value in a column, that entire column will be dropped from the DataFrame. The resulting DataFrame will have all missing values removed along the column axis (axis=1).

```python
nfl_data.dropna()
```

```python
columns_with_na_dropped = nfl_data.dropna(axis=1)
columns_with_na_dropped.shape
```
